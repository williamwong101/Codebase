```python
import matplotlib.pyplot as plt
import numpy as np
```

## Simple Example

```python
x = np.linspace(1,100,10)
y = np.arange(10)*np.random.rand(10)

plt.plot(x, y)
```

## Data structures
- **Figure**
    - The big canvas, contain at least one Axes
- **Axes**
    - The 'plot', can be in only one Figure
- **Axis**
    - Number-line-like objects, take care of setting the graph limits and generating the ticks

```python
fig = plt.figure()  # an empty figure with no Axes
fig, ax = plt.subplots()  # a figure with a single Axes
fig, axs = plt.subplots(2, 2)  # a figure with a 2x2 grid of Axes
```

## Interface style
- object-oriented interface 
- pyplot interface

### OO style

```python
x = np.linspace(0, 3, 100)

fig, ax = plt.subplots()
ax.plot(x, x, label='linear')
ax.plot(x, x**2, label='quadratic')
ax.plot(x, x**3, label='cubic')
ax.set_xlabel("x label")
ax.set_ylabel("y label")
ax.set_title("OO Example")
ax.legend()
```

### Pyplot style

```python
x = np.linspace(0, 3, 100)

plt.plot(x, x, label='linear')  # Plot some data on the (implicit) axes.
plt.plot(x, x**2, label='quadratic')  # etc.
plt.plot(x, x**3, label='cubic')
plt.xlabel('x label')
plt.ylabel('y label')
plt.title("Pyplot Example")
plt.legend()
```
