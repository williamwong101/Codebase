## Indexing
### Basics
- Series: series[label]: return value
- DataFrame: df[colname(s)]: return Series/ DataFrame corresponding to colname(s)

### label based indexing
valid inputs:
- single label 5
- array of labels [5, 'a', 'c']
- slice object 'a':'c'
- boolean array
- callable function

```python
## indexing
# .loc[index, column]: label based
data_df.loc[11410: 17285, 'userID': 'created_at']
```

### position based indexing
valid inputs:
- integer 5
- array of integers [1,2,3,0]
- slice object 1:7
- boolean array
- callable function

```python
data_df.iloc[:10, [2,3,4]]
```

### callable indexing 
- pass in function with one argument, return valid output for indexing
- can be used in label or position based

```python
data_df.loc[lambda df: df['context_annotations'].notnull(), lambda df: ['text', 'urls', 'context_annotations']]
```

## Concatenate, Merge, Join

```python
text = data_df['text']
user = data_df[['username', 'userID']]
data_rows = data_df[:500]
data_row = data_df.iloc[-1,:]
```

### Concatenate DataFrames/ Series

```python
## Concate columns
pd.concat([text, user], axis=1)
```

```python
## Concate rows (append)
# single row
data_rows.append(data_row)
# multiple rows
pd.concat([data_rows, data_rows], axis=0, ignore_index=True)
```
